```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

data_path = '/home/caijb/Desktop/inferset/Nonlang'
seen = ['66','68','70','71','72','76','81','93','100','103']
unseen=['80','87','90','105']
```

```python
error_rate=[]
for i in seen:
    original_path = data_path + "/original/" + i+'.wav'
    gen_path = data_path + '/multi/' + 'seen_'+i+'_griffin.wav'
    y,sr =librosa.load(original_path)
    y_hat, sr_hat = librosa.load(gen_path)
    f0, voiced_flag, voiced_probs = librosa.pyin(y,fmin = 50,fmax=4000)
    times = librosa.times_like(f0)
    f0_hat, voiced_flag_hat, voiced_probs_hat = librosa.pyin(y_hat, fmin = 50,fmax=4000)
    times_hat = librosa.times_like(f0_hat)
    fig, ax =plt.subplots()
    ax.plot(times,f0, label='f0',color='blue',linewidth =2)
    ax.plot(times_hat, f0_hat, label ='f0_hat', color = 'red', linewidth=2)
    f0 = f0[np.isfinite(f0)]
    f0 = np.delete(f0, np.where(f0==50.))
    f0_hat = f0_hat[np.isfinite(f0_hat)]
    f0_hat = np.delete(f0_hat, np.where(f0_hat==50.))
    f0_mean = np.mean(f0)
    f0_hat_mean =np.mean(f0_hat)
    print(f0_mean,f0_hat_mean)
    error_rate.append(abs(f0_mean-f0_hat_mean)/f0_mean)
print(error_rate)
```

```python
print(sum(error_rate)/len(error_rate)*100,"%")
```
